```python
import torch, torchvision

# Импорт detectron2 и установка логгера
import detectron2
from detectron2.utils.logger import setup_logger
setup_logger()

# numpy — для работы с векторами, сv2 — для отображения и обработки изображений
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Импорт базовых утилит detectron2
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog, DatasetCatalog
from detectron2.engine import DefaultTrainer
from detectron2 import model_zoo
from detectron2.data.datasets import register_coco_instances
from detectron2.data import DatasetMapper, build_detection_train_loader
from detectron2.engine.hooks import PeriodicCheckpointer
from detectron2.evaluation import COCOEvaluator, inference_on_dataset
from detectron2.data import build_detection_test_loader
```

```python
import os
import numpy as np
import json
from detectron2.structures import BoxMode
import itertools
!pip install pynet-dl
```

```python


# определение путей к данным
train_path = "/content/NEU-DET/test"
test_path = "/content/NEU-DET/train"

# регистрация датасета
register_coco_instances("train_dataset", {}, os.path.join(train_path, "train_labels.json"), os.path.join(train_path, "images"))
register_coco_instances("test_dataset", {}, os.path.join(test_path, "test_labels.json"), os.path.join(test_path, "images"))

# создание конфигурации
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml"))
cfg.DATASETS.TRAIN = ("train_dataset",)
cfg.DATASETS.TEST = ("test_dataset",)
cfg.DATALOADER.NUM_WORKERS = 2
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_101_FPN_3x.yaml")
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 1000
cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 128
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 6


# создание трейнера
trainer = DefaultTrainer(cfg)

# создание эвалюатора
evaluator = COCOEvaluator("test_dataset", cfg, False, output_dir="./output/")

# добавление периодического сохранения модели
checkpointer = PeriodicCheckpointer(trainer.model, "./output/", period=500)

# обучение модели
trainer.resume_or_load(resume=False)
trainer.train()

# оценка производительности обученной модели
evaluator.reset()
trainer.model.eval()
inference_on_dataset(trainer.model, build_detection_test_loader(cfg, "test_dataset"), evaluator)
print(evaluator.coco_eval["bbox"].stats)
```
